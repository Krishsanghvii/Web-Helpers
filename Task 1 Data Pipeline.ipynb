```python
import requests
from bs4 import BeautifulSoup

# I used beautiful soup as it can scrape data from HTML and XML sites, It being a python library makes it convenient to use for web scraping.

url = 'https://www.amazon.com/Razer-Nari-Ultimate-HyperSense-Gel-Infused/dp/B07HZ6YWNB/ref=sr_1_1_sspa?keywords=gaming+headsets&pd_rd_r=be276a2b-197a-4799-a23b-7331aa832484&pd_rd_w=JutOq&pd_rd_wg=KAxQv&pf_rd_p=12129333-2117-4490-9c17-6d31baf0582a&pf_rd_r=YBEBBYWVPWKJS2XM7VEY&qid=1688751756&sr=8-1-spons&sp_csd=d2lkZ2V0TmFtZT1zcF9hdGY&psc=1'


headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3'
}

response = requests.get(url , headers=headers)

soup = BeautifulSoup(response.text , 'html.parser')

try:
    title = soup.find("span", attrs={"id":'productTitle'}).string.strip()
except:
    title = "N/A"

try:
    category = soup.find("a", attrs={"class":'a-link-normal a-color-tertiary'}).string.strip()
except:
    category = "N/A"

price_span = soup.find('span', {'class': 'a-price aok-align-center reinventPricePriceToPayMargin priceToPay'})
if price_span:
    price = price_span.find('span', {'class': None}).text.strip()
else:
    price = "N/A"

try:
    rating = soup.find('i', attrs={'class': 'a-icon a-icon-star a-star-4-5'}).string.strip()
except:
    rating = "N/A"

try:
    review_count = soup.find('span', attrs={'id': 'acrCustomerReviewText'}).string.strip()
except:
    review_count = "N/A"

try:
    availability = soup.find('div', attrs={'id': 'availability'}).find('span').string.strip()
except:
    availability = "N/A"

try:
    description = soup.find('div', attrs={'id': 'productDescription'}).find('p').string.strip()
except:
    description = "N/A"

print(f'The {title} is under {category} and costs: {price}')
print(f'Rating: {rating}, Reviews: {review_count}, Availability: {availability}')
print(f'Description: {description}')

with open('scraped_products.csv', mode='w', newline='') as file:
    writer = csv.writer(file)
    writer.writerow(['Title', 'Category', 'Price', 'Rating', 'Review Count', 'Availability', 'Description'])
    writer.writerow([title, category, price, rating, review_count, availability, description])

print("Data saved to scraped_products.csv")
```
